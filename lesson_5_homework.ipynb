```python
retail_5 = retail.query('Country == "Germany"').groupby('CustomerID',as_index = False).agg({'InvoiceNo':'nunique'})
```

```python
quan_80 = retail_5.InvoiceNo.quantile(0.8)
```

```python
retail_5 = retail_5.query('InvoiceNo > @quan_80').reset_index(drop = True).sort_values('InvoiceNo',ascending=False)
```

```python
retail_5
```

```python
germany_top = retail_5['CustomerID']
```

```python
germany_top
```

6

```python
top_retail_germany = retail.query('CustomerID in @germany_top')
```
