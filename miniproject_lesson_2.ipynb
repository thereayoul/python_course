```python
def rename_str(st):
    st = st.lower().replace(' ','_')
    return st
```

```python
df = df.rename(columns=rename_str)
```

```python
df.query('is_canceled == 1').groupby(['arrival_date_year',"arrival_date_month"]).agg({'is_canceled':'count'})#.sort_values('is_canceled',ascending = False)
```

```python
df.query('is_canceled == 0').groupby('country').agg({'is_canceled':'count'}).sort_values('is_canceled',ascending=False)
#Пользователи из каких стран совершили наибольшее число успешных бронирований? 
#Бронирование считается успешным, если в дальнейшем не было отменено
```
