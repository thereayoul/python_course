```python
import requests
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
r = requests.get('https://collectionapi.metmuseum.org/public/collection/v1/search?q=sunflowers')
res = r.json()
```

```python
sunflower_ids = res['objectIDs']
```

```python
df = pd.DataFrame()
objectID = []
title = []
artistDisplayName = []
department = []
objectBeginDate = []
objectEndDate = []
period = []
objectName = []
culture = []
```

```python
for i in sunflower_ids:
    try:
        r_test = requests.get(f'https://collectionapi.metmuseum.org/public/collection/v1/objects/' + str(i))
        obj_test = r_test.json()
        objectID.append(obj_test['objectID'])
        title.append(obj_test['title'])
        artistDisplayName.append(obj_test['artistDisplayName'])
        department.append(obj_test['department'])
        objectBeginDate.append(obj_test['objectBeginDate'])
        objectEndDate.append(obj_test['objectEndDate'])
        period.append(obj_test['period'])
        objectName.append(obj_test['objectName'])
        culture.append(obj_test['culture'])
    except:
        continue
```

```python
df['objectID'] = objectID
df['title'] = title
df['artistDisplayName'] = artistDisplayName
df['department'] = department
df['objectBeginDate'] = objectBeginDate
df['objectEndDate'] = objectEndDate
df['period'] = period
df['objectName'] = objectName
df['culture'] = culture
```

```python
r = requests.get('https://collectionapi.metmuseum.org/public/collection/v1/search?departmentId=6&q=cat')
cats = r.json()
```

```python
cat_ids = cats['objectIDs'][:1000]
```

```python
df_cat['objectID'] = objectID
df_cat['title'] = title
df_cat['artistDisplayName'] = artistDisplayName
df_cat['department'] = department
df_cat['objectBeginDate'] = objectBeginDate
df_cat['objectEndDate'] = objectEndDate
df_cat['period'] = period
df_cat['objectName'] = objectName
df_cat['culture'] = culture
```
