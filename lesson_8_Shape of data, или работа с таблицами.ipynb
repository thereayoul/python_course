```python
import pandas as pd
```

# 3

```python
pokemon = pokemon.rename(columns={'#':'id'}).rename(columns=(lambda x: x.replace(' ','_').replace('.','').lower()))
```

```python
pokemon.head(1)
```

```python
legends = pokemon.groupby('generation').legendary.value_counts().to_frame()
```

```python
legends
```

# 4

```python
legends = legends.rename(columns={'legendary':'legendary_count'})
```

```python
legends_unstacked = legends.unstack('legendary')
```

```python
legends_unstacked
```

```python
(pokemon.groupby(['generation','type_1'])
    .legendary.value_counts()
    .to_frame()
    .unstack()
    .loc[:,('legendary', True)]
    .idxmax()
)
```

# 7

```python
avocado_agg = pd.DataFrame({'type' : ['conventional', 'organic'],
                            'AvgPrice_2015' : [1.077963, 1.673324],
                            'AvgPrice_2016' : [1.105595, 1.571684],
                            'AvgPrice_2017' : [1.294888, 1.735521],
                            'AvgPrice_2018' : [1.127886, 1.567176],
                            })
avocado_agg
```

```python
avocado_agg_long = pd.wide_to_long(avocado_agg, ['AvgPrice'], i = 'type', j ='year', sep='_')
avocado_agg_long
```

```python
superheroes = pd.read_csv('superheroes_power_matrix.csv')
superheroes
```

```python
superheroes_long = superheroes.melt(id_vars='Name', var_name='superpower')
superheroes_long
```

```python
superheroes_powers = superheroes_long.query('value == True').groupby('Name').superpower.apply(list).reset_index()
```
